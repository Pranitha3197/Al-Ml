```python
import pandas as pd
```

```python

df=pd.read_csv("/content/breast_cancer_survival.csv")
```

```python
df
```

```python
df['ER status']=df['ER status'].replace({'Negative':0,'Positive':1})
df['PR status']=df['PR status'].replace({'Negative':0,'Positive':1})
df['HER2 status']=df['HER2 status'].replace({'Negative':0,'Positive':1})
df['Gender']=df['Gender'].replace({'MALE':0,'FEMALE':1})
df['Patient_Status']=df['Patient_Status'].replace({'Dead':0,'Alive':1})
df['Histology']=df['Histology'].replace({'Infiltrating Ductal Carcinoma':0,'Infiltrating Lobular Carcinoma':1,'Mucinous Carcinoma':2})
df['Tumour_Stage']=df['Tumour_Stage'].replace({'II':0,'I':1,'III':2})
df['Surgery_type']=df['Surgery_type'].replace({'Lumpectomy':0,'Modified Radical Mastectomy':1,'Simple Mastectomy':2,'Other':3})

```

```python
df
```

```python


df.drop('Date_of_Surgery',inplace=True,axis=1)



df.drop('Date_of_Last_Visit',inplace=True,axis=1)


df
```

```python
y=df['Patient_Status']
y
```

```python
df.fillna(0,inplace=True)
df
```

```python
x=df.drop('Patient_Status',axis=1)
x
```

```python
columns_to_convert = ['Protein1', 'Protein2', 'Protein3','Protein4','Patient_Status']

df[columns_to_convert] = df[columns_to_convert].astype(int)


print(df.dtypes)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn import svm
from sklearn.svm import SVC
svc=SVC()
svc.fit(x_train,y_train)
y_p=svc.predict(x_test)
y_p
```

```python

from sklearn.metrics import accuracy_score
accuracy=accuracy_score(y_test,y_p)
```

```python
accuracy
```

```python
from sklearn.metrics import classification_report
report = classification_report(y_test, y_p)
print(report)
```

```python

from sklearn.metrics import confusion_matrix
confusion = confusion_matrix(y_test, y_p)
print(confusion)
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report, confusion_matrix
import numpy as np
```

```python
param_grid = {
    'C': [0.1, 1, 10, 100, 1000],
    'gamma': [1, 0.1, 0.01, 0.001, 0.0001],
    'kernel': ['rbf']
}
```

```python
svc = SVC()

grid = GridSearchCV(estimator=svc, param_grid=param_grid, refit=True, verbose=2)
grid.fit(x_train, y_train)
print("Best Parameters:", grid.best_params_)
print("Best Score:", grid.best_score_)
```

```python
y_p = grid.predict(x_test)
```

```python

print("Classification Report:")
print(classification_report(y_test, y_p))
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_p))
```
